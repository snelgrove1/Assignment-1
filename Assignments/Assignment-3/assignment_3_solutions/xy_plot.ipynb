```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (a) loading the data from xy_points.dat
file = open('xy_points.dat', mode="r")

# first line contains x values
temp = []
line = file.readline()
for v in line.split():
    temp.append(float(v))
# converting to numpy array    
x = np.array(temp)

# second line contains y values
temp = []
line = file.readline()
for v in line.split():
    temp.append(float(v))
# converting to numpy array    
y = np.array(temp)

# closing the file as soon as we finished with it
file.close()

# (e) customising the axes to have square region
plt.figure(figsize=(15,15), dpi=96)

# (b) ploting all the (x, y) coordinate pairs as isolated red dots 
plt.plot(x, y,'ro')

# (c) using logical indexing to plot all the pairs that satisfy
# the inequalities (x < −2) and (−5 <= y <= 0) as blue dots.
# note that np.where(cond) returns index to the entires in the
# array where 'cond' is true.
logical_index = np.where(np.logical_and(x < -2, np.logical_and(y >= -5, y <=0)))
plt.plot(x[logical_index], y[logical_index],'bo')

# (d) using logical indexing to plot all the entries that are
# within radial distance of 5 from x=0,y=5 as green dots  
radial_dist = np.empty(len(x))
for i in range(len(x)):
    radial_dist[i] = ((x[i] - 0)**2 + (y[i] - 5)**2)**0.5
plt.plot(x[radial_dist <= 5], y[radial_dist <= 5],'go')

# (e) adding title to the plot
plt.title("Highlighted points (100XXXXXX)", fontsize=20)

# showing the final plot
plt.show()
```
