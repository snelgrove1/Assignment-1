# Vaccine Map

## Converting Excel File

`Source_data_for_CFR_vaccine_map_abridged.xlsx` was opened in MS Excel. Columns A, E, F, and I were selected by holding `Ctrl` button then `Ctrl-C` (copy to clipboard), followed by `Ctrl-N` (new workbook) and finally `Ctrl-V` (paste clipboard). Because there are 6 items in column I (i.e. Cases) that have commas as 1000 seperator (i.e. rows: #188, #195, #221, #225, #377 and #495); first, column I was selected then `right-click > Format Cells...` and under `Category`, `Number` was selected. In `Number` dialog, `Decimal places` was set to `0` and `Use 1000 Seperator (,)` unchecked and `OK` button clicked. At the end, the file was `Saved As` a `CSV` file. That's all there's to it!

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# setting structured data type for numpy
dt = np.dtype([('category', 'U15'), ('lat', float), ('long', float), ('cases', float)])

# using numpy's recfromcsv() to read from csv file
data = np.recfromcsv('Source_data_for_CFR_vaccine_map_abridged.csv', delimiter=',', dtype=dt)

# finding largest measles outbreak
max = 0
for i in range(len(data)):
    if (data['category'][i] == 'Measles' and data['cases'][i] > max):
        max = data['cases'][i]

# spherical scale (^1/3) the radii of the other
# disks as fractions of the largest radius
for i in range(len(data)):
        data['cases'][i] = (data['cases'][i] / max)**(1/3) * 28

# using basemap to plot
fig = plt.figure(figsize=(14,10))
ax = plt.subplot(1,1,1)
earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.93', lake_color=(0.757, 0.824, 0.824))
earth.drawstates(color='0.50', linewidth=0.1)
earth.drawcountries(color='0.50', linewidth=0.25)
earth.drawmapboundary(fill_color=(0.757, 0.824, 0.824))

for i in range (len(data)):
    if data['category'][i] == 'Measles':
        # using earth object to translate coordinates automatically
        x, y = earth(data['long'][i], data['lat'][i])
        earth.plot(x, y, 'ro', markersize=data['cases'][i], alpha=0.45)    
    
plt.title("Measles outbreaks (Student ID)")
plt.show()
```
